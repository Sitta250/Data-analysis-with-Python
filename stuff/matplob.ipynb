```python
salary_data = df['salary_year_avg'].dropna()
salary_data
```

```python
import matplotlib.ticker as mticker
plt.hist(salary_data, bins = 30, edgecolor='black')
plt.gca().xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x/1000)}K'))
plt.ylabel('job count')
plt.xlabel('yearly salary')
```

```python
analyst_salary = df[df['job_title_short']== 'Data Analyst']['salary_year_avg'].dropna()
engineer_salary = df[df['job_title_short']== 'Data Engineer']['salary_year_avg'].dropna()
```

```python
plt.hist([analyst_salary, engineer_salary], bins=20, alpha=0.5, label=['data analyst', 'data engineer'])
plt.gca().xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x/1000)}K'))
plt.title('Salary Distribution: Data Analyst vs Data Engineer')
plt.xlabel('Yearly Salary (in K)')
plt.ylabel('Frequency')
plt.legend()
```

```python
df_clean = df[['job_posted_date', 'salary_year_avg']].dropna()
df_clean['job_posted_date'] = pd.to_datetime(df_clean['job_posted_date'])
df_clean = df_clean.set_index('job_posted_date').resample('M').mean()
```

```python
df_clean.plot(y='salary_year_avg', kind = 'line', figsize=(10,5))
plt.title("Average Salary Over Time")
plt.xlabel("Date")
plt.ylabel("Avg Salary")
plt.grid(True)
plt.show()
```

```python
df_remote_overtime = df[df['job_work_from_home']==True].groupby('job_posted_month').size()
```

```python
df_remote_overtime.plot(kind = 'line', marker = 'o', figsize=(10,5))
plt.title('remote job posted per month')
plt.xlabel('month')
plt.ylabel('numbers of jobs posted')
plt.grid=(True)
plt.show()
```

```python
df_job_med = df.groupby('job_title_short')['salary_year_avg'].agg('median').sort_values(ascending=True)
```
