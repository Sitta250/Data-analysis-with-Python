```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
def remote_job_percentage(df):
    total_jobs = len(df)  # Count total job listings
    remote_jobs = df['job_work_from_home'].sum()  # Sum True values (since True = 1, False = 0)
    return (remote_jobs / total_jobs) * 100  # Calculate percentage

remote_percent = remote_job_percentage(df)
print(f"Remote Job Percentage: {remote_percent:.2f}%")
```
