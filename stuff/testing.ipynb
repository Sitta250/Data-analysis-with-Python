```python
from datetime import datetime
def convert_datetime(date_str):
  return datetime.strptime(date_str, '%Y-%m-%d')

print(convert_datetime('2023-06-15'))
```

```python
def cal_age(date_str):
  birth_date = datetime.strptime(date_str, '%Y-%m-%d')
  current_year = datetime.today().year
  return current_year-birth_date.year

print(cal_age('2000-01-01'))
```

```python
file = open('california_housing_test.csv')
content = file.read()
file.close()
```

```python
import csv
data_dict = {}
for index, row in enumerate(csv.reader(content.strip().split('\n'))):
  if index ==0:
    #initialize dict with column headers as keys
    for column in row:
      data_dict[column] = []
  else:
    #append each element in the row to the correct list in the dict
    for col_index, value in enumerate(row):
      data_dict[list(data_dict.keys())[col_index]].append(value)
```

```python
sum(float(room.replace(' ',''))for room in data_dict['total_rooms'] if room.replace(' ', '').replace('.','').isdigit())
```

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=0.1, symbol="$"):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol= symbol
    self.total_salary = base_salary*(1+bonus_rate)
    self.bonus = self.total_salary-base_salary
  def __repr__(self):
    return f'{self.symbol}{self.base_salary:,.0f}'
  def calculate_salary(self):
    return f'{self.symbol}{self.total_salary:,.0f}'
  def calculate_bonus(self):
    return f'{self.symbol}{self.bonus:,.0f}'
```

```python
salary = BaseSalary(100000)
```

```python
salary.calculate_bonus()
```
